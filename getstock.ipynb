```python
import unittest
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
import time
import csv
# import cStringIO, codecs

class WebTable:
    def __init__(self, webtable):
       self.table = webtable

    def get_row_count(self):
        return len(self.table.find_elements_by_tag_name("tr")) - 1

    def get_column_count(self):
        return len(self.table.find_elements_by_xpath("//tbody/tr[2]/td"))

    def get_table_size(self):
        return {"rows": self.get_row_count(),
                "columns": self.get_column_count()}

    def row_data(self, row_number):
        if(row_number == 0):
            raise Exception("Row number starts from 1")
        row_number = row_number + 1
        row = self.table.find_elements_by_xpath("//tr["+str(row_number)+"]/td")
        rData = []
        for webElement in row :
            if webElement.text != "":
                rData.append(webElement.text)
        return rData

    def column_data(self, column_number):
        col = self.table.find_elements_by_xpath("//tr/td["+str(column_number)+"]")
        rData = []
        for webElement in col :
            rData.append(webElement.text)
        return rData

    def get_all_data(self):
        # get number of rows
        noOfRows = len(self.table.find_elements_by_xpath("//tr")) -1
        # get number of columns
        # noOfColumns = len(self.table.find_elements_by_xpath("//tr[2]/td"))
        allData = []
        # iterate over the rows, to ignore the headers we have started the i with '1'
        for i in range(2, noOfRows):
            # reset the row data every time
            ro = []
            # noOfColumns = len(self.table.find_elements_by_xpath("//tr["+str(i)+"]/td"))
            all_fields =  self.table.find_elements_by_xpath("//tr["+str(i)+"]/td")
            # iterate over columns
            for field in all_fields :
                # get text from the i th row and j th column
                if field.text != "":
                    ro.append(field.text)
            # add the row data to allData of the self.table
            allData.append(ro)
        return allData


class Test(object):
    # def __init__(self, table):
    #    self.table = t

    def test_web_table(self):
        outfile = open('table9.csv','w',newline='', encoding ='utf-8')
        writer = csv.writer(outfile)    
        driver = webdriver.Chrome(executable_path=r'c:\chromedriver.exe')
        # driver.maximize_window()
        driver.implicitly_wait(30)
        driver.get("https://www.stockbiz.vn/IndicesStats.aspx")
        time.sleep(3)
        WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//button[@class='mfp-close']"))).click()
        
        for j in range(1,5):       
            if j > 1 :
                WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//div[@class='pageNavigation']/a[contains(@onclick,'RefreshData("+ str(j) +")')]"))).click()
                WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//table[@class='dataTable']"))) 
                time.sleep(3)
            w = WebTable(driver.find_element_by_xpath("//table[@class='dataTable']/tbody"))   
            row = []
            for i in range(1,w.get_row_count()+1):
                    print("Row " +str(i) +" data : ", w.row_data(i))
                    row.append(w.row_data(i))
                    # WebDriverWait(driver, 60).until(EC.presence_of_element_located((By.XPATH, "//table[@class='dataTable']/tbody")))        
            #         print("\n")
            for item in row:
                writer.writerow(item)
        # print(' '.join(item))  
        outfile.close()   
        driver.close()
        driver.quit()

if __name__ == "__main__":
    # unittest.main()        
    Test().test_web_table()
```
